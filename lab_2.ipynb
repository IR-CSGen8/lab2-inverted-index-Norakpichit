```python
# Import necessary libraries
import pandas as pd
import json

# Load the CSV dataset
df = pd.read_csv("semi_strut.csv")

# Display the first few rows of the dataset
df.head()
```

```python
# Tokenization function to extract terms from the JSON-like content
# Remember to exact both 
# 1 .Extract terms from various fields (title, keywords, and content)
values = df['Content'].values

# 2. Extract terms from sections' titles and content


# 5. you can display the DataFrame
# def tokenize_content(content):
```

```python
values = df['Content'].values
def column_to_terms(column):
    dictionaries = []
    # convert numpy array to list of dictionary
    for value in column:
        dictionaries.append(json.loads(value))
    
    unique_terms = set()
    
    for dictionary in dictionaries:
        # Find unique terms in 'title'
        for term_title in dictionary['title'].split():
            unique_terms.add(term_title)
        # Find unique terns in 'keyword'
        for term_keyword in dictionary['keywords']:
            unique_terms.add(term_keyword)
        for section in dictionary['sections']:
            for term_section_title in section['title'].split():
                unique_terms.add(term_section_title)
            for term_content in section['content'].split():
                unique_terms.add(term_content)
    
    return unique_terms
```

```python
unique_terms = column_to_terms(values);
unique_terms
```

```python
# 3. Tokenize the content and create a new column "Terms"
# sort the unique terms set
unique_terms = sorted(unique_terms)
# convert the set to a new table
terms_df = pd.DataFrame(unique_terms, columns=["Terms"])
terms_df
```

```python
# Initialize an empty inverted index dictionary the extracted terms 
df.head()
def inverted_index(terms):
    inverted_index = {}
    for term in terms:
        inverted_index[term] = set()
        for i, dictionary in enumerate(dictionaries):
            if term in dictionary['title']:
                inverted_index[term].add(i+1)
            if term in dictionary['keywords']:
                inverted_index[term].add(i+1)
            for section in dictionary['sections']:
                if term in section['title']:
                    inverted_index[term].add(i+1)
                if term in section['content']:
                     inverted_index[term].add(i+1)
    return inverted_index
            
inverted_index(unique_terms)
```

```python
# 4. Implement a preprocessing function that converts terms to lowercase, removes punctuation, and removes common stop words.
    # Create another new column "Terms_preprocessed"
def preprocess(terms):
    lowercase_terms = set()
    for term in unique_terms:
        lowercase_terms.add(term.lower())
    
    rp_terms = set()
    for term in lowercase_terms:
        temp_str = ''.join(letter for letter in term if letter.isalnum())
        rp_terms.add(temp_str)
    
    clean_terms = set()
    common_words = ["a", "an", "the", "this", "that", "is", "it", "to", "and", "are", "in", "of", "for", "be", "can"]
    for term in rp_terms:
        if term in common_words:
            continue
        else:
            clean_terms.add(term)
    clean_terms = sorted(clean_terms)
    return clean_terms

preprocess(unique_terms)
```
